```python
# IMPORT LIBRARIES
# import all the necessary libraries
import pandas as pd
import numpy as np

# data cleaning
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# data visualization
import matplotlib.pyplot as plt

# cross validation
import statsmodels.api as sim
from statsmodels.tsa.stattools import acf, pacf

# Model training and prediction
from statsmodels.tsa.arima.model import ARIMA


# Import the data 
filepath = 'C:/Users/ernest/Documents/Helen_NAPTIN/'
filename = 'dataset_for_rice_and_maize.xlsx'
dataframe = pd.read_excel(filepath + filename)
dataframe.head(5)


# DATA EXPLORATION
# shape of the data
shape = dataframe.shape
print(f'Shape of data is: ', shape)

# check for missing values
dataframe.isna().sum()

# the datatype in the dataset
types = dataframe.dtypes
print('\n', types)

#convert the date column to datetime
dataframe['ItemLabels'] =  pd.to_datetime(dataframe['ItemLabels'])
print('\n',dataframe.head())

#set the date column (ItemLabels) to index
dataframe.set_index('ItemLabels', inplace=True)

# rename the dataframe
ts = dataframe
print('\nData statistical Analysis\n',ts.describe().T, '\n')

# plot
fig, ax=plt.subplots(figsize=(16, 7))
ts['Rice Price'].plot(ax=ax, label='Rice')
ts['Maize Price'].plot(ax=ax, label='Maize')
ax.legend();

ts_rice = ts['Rice Price']
ts_maize = ts['Maize Price']
ts_rice.head()
ts_maize.head()

fig, ax=plt.subplots(figsize=(16, 7))
ts_rice.plot(ax=ax, label='Rice')
ax.legend();

fig, ax=plt.subplots(figsize=(16, 7))
ts_maize.plot(ax=ax, label='Maize', color='red')
ax.legend();


# DATA ANALYSIS
# Rice ADF Test
rice_result = adfuller(ts_rice, autolag='AIC')
print(f'Rice ADF Statistic: {rice_result[0]}')
print(f'n_lags: {rice_result[1]}')
print(f'p-value: {rice_result[1]}')
for key, value in rice_result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
    
# Maize ADF Test
maize_result = adfuller(ts_maize, autolag='AIC')
print(f'\nMaize ADF Statistic: {maize_result[0]}')
print(f'n_lags: {maize_result[1]}')
print(f'p-value: {maize_result[1]}')
for key, value in maize_result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
    
# STATIONARITY   
# rice - rolling stat
roll_mean_rice = ts_rice.rolling(12).mean()
roll_std_rice = ts_rice.rolling(12).std()

# plot rice rolling stat
plt.figure(figsize = (16,7))
fig = plt.figure(1)
original = plt.plot(ts_rice, color='blue', label='Original')
mean = plt.plot(roll_mean_rice, color='red', label='Rolling Mean')
std = plt.plot(roll_std_rice, color='black', label='Rolling std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation for Rice')
plt.show(block = False)

# maize - rolling stat
roll_mean_maize = ts_maize.rolling(12).mean()
roll_std_maize = ts_maize.rolling(12).std()

# plot maize rolling stat 
plt.figure(figsize = (16,7))
fig = plt.figure(1)
original = plt.plot(ts_maize, color='blue', label='Original')
mean = plt.plot(roll_mean_maize, color='red', label='Rolling Mean')
std = plt.plot(roll_std_maize, color='black', label='Rolling std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation for Maize')
plt.show(block = False)


# TRANSFORMATION
# Rice - log transformation
plt.figure(figsize = (16,7))
fig = plt.figure(1)
ts_rice_log = np.log(ts_rice)
plt.plot(ts_rice_log)

# Maize - log transformation
plt.figure(figsize = (16,7))
fig = plt.figure(1)
ts_maize_log = np.log(ts_maize)
plt.plot(ts_maize_log)


# DECOMPOSITION
# Rice Decomposition
rice_decomposition = seasonal_decompose(ts_rice_log, freq=1, model = 'multiplicative')

trend = rice_decomposition.trend
seasonal = rice_decomposition.seasonal
residual = rice_decomposition.resid

plt.figure(figsize = (16,7))
fig = plt.figure(1)

plt.subplot(411)
plt.plot(ts_rice_log, label='Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(ts_rice_log, label='Trend', color='red')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(ts_rice_log, label='Seasonality', color='green')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(ts_rice_log, label='Residuals', color='black')
plt.legend(loc='best')

# Maize Decomposition

maize_decomposition = seasonal_decompose(ts_rice_log, freq=1, model = 'multiplicative')

trend = maize_decomposition.trend
seasonal = maize_decomposition.seasonal
residual = maize_decomposition.resid

plt.figure(figsize = (16,7))
fig = plt.figure(1)

plt.subplot(411)
plt.plot(ts_maize_log, label='Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(ts_maize_log, label='Trend', color='red')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(ts_maize_log, label='Seasonality', color='green')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(ts_maize_log, label='Residuals', color='black')
plt.legend(loc='best')


# Rice differencing
plt.figure(figsize=(16,7))
fig = plt.figure(1)
ts_rice_log_diff = ts_rice_log - ts_rice_log.shift()
plt.plot(ts_rice_log_diff)

# Maize differencing
plt.figure(figsize=(16,7))
fig = plt.figure(1)
ts_maize_log_diff = ts_rice_log - ts_maize_log.shift()
plt.plot(ts_maize_log_diff)

# TRANSFORMED
# rice transformed rolling stat
roll_mean_rice_diff = ts_rice.rolling(12).mean()
roll_std_rice_diff = ts_rice.rolling(12).std()

# plot rice transformation rolling stat 
plt.figure(figsize = (16,7))
fig = plt.figure(1)
original = plt.plot(ts_rice_log_diff, color='blue', label='Original')
rice_mean = plt.plot(roll_mean_rice_diff, color='red', label='Rolling Mean')
rice_std = plt.plot(roll_std_rice_diff, color='black', label='Rolling std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation for Rice after cleaning')
plt.show(block = False)


# maize transformed rolling stat
roll_mean_maize_diff = ts_maize.rolling(12).mean()
roll_std_maize_diff = ts_maize.rolling(12).std()

# plot rolling stat
plt.figure(figsize = (16,7))
fig = plt.figure(1)
original = plt.plot(ts_maize_log_diff, color='blue', label='Original')
maize_mean = plt.plot(roll_mean_maize_diff, color='red', label='Rolling Mean')
maize_std = plt.plot(roll_std_maize_diff, color='black', label='Rolling std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation for Maize after cleaning')
plt.show(block = False)

# CORRELATION
#rice cross validation
lag_acf = acf(ts_rice_log_diff, nlags = 20)
lag_pacf = pacf(ts_rice_log_diff, nlags = 20)
fig = plt.figure(figsize=(16,7))
ax1 = fig.add_subplot(211)
fig = sim.graphics.tsa.plot_acf(ts_rice_log_diff.dropna(), ax = ax1)
ax2 = fig.add_subplot(212)
fig = sim.graphics.tsa.plot_pacf(ts_rice_log_diff.dropna(), ax = ax2)

# Maize cross validation
lag_acf = acf(ts_maize_log_diff, nlags = 20)
lag_pacf = pacf(ts_maize_log_diff, nlags = 20)
fig = plt.figure(figsize=(16,7))
ax1 = fig.add_subplot(211)
fig = sim.graphics.tsa.plot_acf(ts_maize_log_diff.dropna(), ax = ax1)
ax2 = fig.add_subplot(212)
fig = sim.graphics.tsa.plot_pacf(ts_maize_log_diff.dropna(), ax = ax2)


# TRAINING
# fit model for rice
plt.figure(figsize=(16,7))
model = ARIMA(ts_rice_log_diff, order = (1,0,0))
rice_results_ARIMA = model.fit()
plt.plot(ts_rice_log_diff)
plt.plot(rice_results_ARIMA.fittedvalues, color='red')

# fit model for maize
plt.figure(figsize=(16,7))
model = ARIMA(ts_maize_log_diff, order = (1,0,0))
maize_results_ARIMA = model.fit()
plt.plot(ts_maize_log_diff)
plt.plot(maize_results_ARIMA.fittedvalues, color='red')

# RETURN TO SCALE
# rice return to original scale
ARIMA_diff_pred = pd.Series(rice_results_ARIMA.fittedvalues, copy=True)
print(ARIMA_diff_pred.head())

ARIMA_diff_pred_cumsum = ARIMA_diff_pred.cumsum()
print(ARIMA_diff_pred_cumsum.head())

ARIMA_log_pred = pd.Series(ts_rice_log.iloc[0], index=ts_rice_log.index)
ARIMA_log_pred = ARIMA_log_pred.add(ARIMA_diff_pred_cumsum, fill_value=0)
ARIMA_log_pred.head()

# maize return to original scale
ARIMA_diff_pred = pd.Series(maize_results_ARIMA.fittedvalues, copy=True)
print(ARIMA_diff_pred.head())

ARIMA_diff_pred_cumsum = ARIMA_diff_pred.cumsum()
print(ARIMA_diff_pred_cumsum.head())

ARIMA_log_pred = pd.Series(ts_maize_log.iloc[0], index=ts_maize_log.index)
ARIMA_log_pred = ARIMA_log_pred.add(ARIMA_diff_pred_cumsum, fill_value=0)
ARIMA_log_pred.head()

# PREDITION
# rice prediction
plt.figure(figsize=(16,7))
prediction_rice_ARIMA = np.exp(ARIMA_log_pred)
plt.plot(ts_rice, label='Original Rice Price')
plt.plot(prediction_rice_ARIMA, label='Predicted Rice Price')
prediction_rice_ARIMA.plot(ax=ax, label='Predicted Rice Price')
plt.title('Rice RMSE: %.4f!'% np.sqrt(sum((prediction_rice_ARIMA - ts_rice)**2/len(ts_rice))))

# maize prediction
plt.figure(figsize=(16,7))
prediction_maize_ARIMA = np.exp(ARIMA_log_pred)
plt.plot(ts_maize, label='Original Maize Price')
plt.plot(prediction_maize_ARIMA, label='Predicted Maize Price')
prediction_maize_ARIMA.plot(ax=ax, label='Predicted Maize Price')
plt.title('Maize RMSE: %.4f!'% np.sqrt(sum((prediction_maize_ARIMA - ts_maize)**2/len(ts_maize))))

# rice future preditction
ARIMA_future_rice = rice_results_ARIMA.predict('2017-01-01','2024-01-01')

ARIMA_future_rice_pred = pd.Series(ARIMA_future_rice, copy=True)
print(ARIMA_future_rice_pred.tail())

ARIMA_future_rice_pred_cumsum = ARIMA_future_rice_pred.cumsum()
print(ARIMA_future_rice_pred_cumsum.tail())

ARIMA_fut_rice_log_pred = pd.Series(ts_rice_log.iloc[0], index=ts_rice_log.index)
ARIMA_fut_rice_log_pred = ARIMA_fut_rice_log_pred.add(ARIMA_future_rice_pred_cumsum, fill_value=0)
ARIMA_fut_rice_log_pred.tail()

plt.figure(figsize=(16,7))
future_rice_prediction_ARIMA = np.exp(ARIMA_fut_rice_log_pred)
plt.plot(ts_rice, label='Original Rice Price')
plt.plot(future_rice_prediction_ARIMA, label='Predicted Rice Price')
future_rice_prediction_ARIMA.plot(ax=ax, label='Predicted Rice Price')
plt.title('Rice Prediction for the next 27 Months')

# maize future prediction
ARIMA_future_maize = maize_results_ARIMA.predict('2017-01-01','2024-01-01')

ARIMA_future_maize_pred = pd.Series(ARIMA_future_maize, copy=True)
print(ARIMA_future_maize_pred.tail())

ARIMA_future_maize_pred_cumsum = ARIMA_future_maize_pred.cumsum()
print(ARIMA_future_maize_pred_cumsum.tail())

ARIMA_fut_maize_log_pred = pd.Series(ts_maize_log.iloc[0], index=ts_maize_log.index)
ARIMA_fut_maize_log_pred = ARIMA_fut_maize_log_pred.add(ARIMA_future_maize_pred_cumsum, fill_value=0)
ARIMA_fut_maize_log_pred.tail()

plt.figure(figsize=(16,7))
future_maize_prediction_ARIMA = np.exp(ARIMA_fut_maize_log_pred)
plt.plot(ts_maize, label='Original Rice Price')
plt.plot(future_maize_prediction_ARIMA, label='Predicted Rice Price')
future_maize_prediction_ARIMA.plot(ax=ax, label='Predicted Rice Price')
plt.title('Maize Prediction for the next 27 Months')

# fitting uncleaned rice data 
plt.figure(figsize=(16,7))
model = ARIMA(ts_rice, order = (1,0,0))
rice_original_ARIMA = model.fit()
plt.plot(ts_rice)
plt.plot(rice_original_ARIMA.fittedvalues, color='red')

# fitting uncleaned maize data
plt.figure(figsize=(16,7))
model = ARIMA(ts_maize, order = (1,0,0))
maize_original_ARIMA = model.fit()
plt.plot(ts_maize)
plt.plot(maize_original_ARIMA.fittedvalues, color='red')

# predict future uncleaned rice data
ARIMA_ori_rice_fut_results = rice_original_ARIMA.predict('2017-01-01','2024-01-01')
print(ARIMA_ori_rice_fut_results.tail())

plt.figure(figsize=(16,7))
plt.plot(ts_rice, label='Original Rice Price')
plt.plot(ARIMA_ori_rice_fut_results, label='Predicted Rice Price')
ARIMA_ori_rice_fut_results.plot(ax=ax, label='Predicted Rice Price')
plt.title('Rice original Prediction for the next 27 Months')

# predict future uncleaned maize data
ARIMA_ori_maize_fut_results = maize_original_ARIMA.predict('2017-01-01','2024-01-01')
print(ARIMA_ori_maize_fut_results.tail())

plt.figure(figsize=(16,7))
plt.plot(ts_maize, label='Original Maize Price')
plt.plot(ARIMA_ori_maize_fut_results, label='Predicted Maize Price')
ARIMA_ori_maize_fut_results.plot(ax=ax, label='Predicted Maize Price')
plt.title('Maize original Prediction for the next 27 Months')


# COMPARISON
# comparison plot btw the uncleaned rice ts and the cleaned rice ts
plt.figure(figsize=(16,7))
plt.plot(ARIMA_ori_rice_fut_results, label='Predicted Rice Price')
plt.plot(future_rice_prediction_ARIMA, label='Predicted Rice Price')
plt.title('Comparsion of predictions')
plt.title('Rice RMSE: %.4f!'% np.sqrt(sum((future_rice_prediction_ARIMA - ARIMA_ori_rice_fut_results)**2/len(ARIMA_ori_rice_fut_results))))

# comparison plot btw uncleaned maize ts and cleaned maize ts
plt.figure(figsize=(16,7))
plt.plot(ARIMA_ori_maize_fut_results, label='Predicted Maize Price')
plt.plot(future_maize_prediction_ARIMA, label='Predicted Maize Price')
plt.title('Comparsion of predictions')
plt.title('Rice RMSE: %.4f!'% np.sqrt(sum((future_maize_prediction_ARIMA - ARIMA_ori_maize_fut_results)**2/len(ARIMA_ori_maize_fut_results))))
    
```
